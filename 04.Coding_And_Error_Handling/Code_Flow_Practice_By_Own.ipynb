```python
#Scenario-01 : Shopping Items and Discount Calculation

cart_items = ["Laptop", "Mouse", "Keyboard", "Monitor"]
item_prices = {"Laptop": 1000, "Mouse": 20, "Keyboard": 50, "Monitor": 300}

# Calculate the total cost
#First Sum the Item's Values
#Create a New Variable and Store the Summed Value

total_cost = sum(item_prices[item] for item in cart_items)
print("Calculation-01:\nTotal Cost Before Discount :", total_cost)

#Calculate Discount of 10%, if the Total Purcharsed Value is More than 200

discount_rate = 0.1

if total_cost > 200 :
    total_cost -= total_cost*discount_rate
print("\nCalculation-02:\nTotal Cost After Discount :", total_cost)

class Customer_Order:
    def __init__(self,items,totalCost,discountRate):
        self.items=items
        self.total_cost=total_cost
        self.discount_rate=discount_rate
    
    def Order_Summary(self):
        print("\nOrder Summary :")
        print("Total Items :",','.join(self.items))
        print("Total Cost :",self.total_cost)
        print("Total Discount :",self.discount_rate)
        
customer_order = Customer_Order(cart_items, total_cost, discount_rate)
customer_order.Order_Summary()
```

```python
#Scenario-02 : Calculation of Departmentwise Salary and Hike for Specific Department 

#1.Employee Details:
#This dictionary contains information about employees, including their departments and salaries.

employees = {
    "John": {"department": "HR", "salary": 50000},
    "Alice": {"department": "Engineering", "salary": 70000},
    "Bob": {"department": "Marketing", "salary": 60000}
}

#2.Initialize Department Salaries:
#An empty dictionary is created to store the total salaries for each department.

department_salaries = {}


#3.Iterate Through Employees:
#This loop goes through each employee and their details:
#department is extracted from the employee's details.
#salary is extracted from the employee's details.
#The total salary for each department is updated. If the department is not already in department_salaries, it is added with a default value of 0.

for employee, details in employees.items():
    department = details["department"]
    salary = details["salary"]
    department_salaries[department] = department_salaries.get(department,0) + salary
    
#4.Display total salaries for each department
#This prints the total salaries for each department. 
#It iterates through department_salaries and displays the department along with its total salary.

print("\nTotal Salaries by Department:")
for department, total_salary in department_salaries.items():
    print(f"{department}: ${total_salary}")
    

#5.Define the raise percentage (for demonstration, let's give a 10% raise to Engineering employees)

raise_percentage = 0.1

#6.Apply the raise to the relevant employees
#The loop iterates through the employees and checks if the employee belongs to the "Engineering" department.
#If the employee is in the "Engineering" department, their salary is increased by 10%.

for employee, details in employees.items():
    if details["department"] == "Engineering":
        details["salary"] += details["salary"] * raise_percentage

#7.Display employee details after the raise
#The loop iterates through the employees dictionary and prints the updated details for each employee, 
#including their department and new salary.

print("\nEmployee Details After Raise:")
for employee, details in employees.items():
    department = details["department"]
    salary = details["salary"]
    print(f"{employee} - Department: {department}, Salary: ${salary}")
```

```python
#Scenario-03 : Verifying Availability of Book(s) for given Input and Update their Availability Count accordingly 

#1.The library_books dictionary stores the initial details of the available books and their copies.
library_books = {
    "Python Programming": 5,
    "Data Science for Beginners": 3,
    "Classic Novels Collection": 8,
    "Web Development Basics": 6,
    "Introduction to AI": 4
}

# Display available books
print("Available Books:")

#A loop prints the titles and the number of copies available for each book.
for book, copies in library_books.items():
    print(f"{book} - Copies available: {copies}")

# Second code snippet: Simulate a user borrowing a book
#The user is prompted to enter the title of the book they want to borrow (user_book).

user_book = input("\nEnter the book you want to borrow: ")

# Check if the book is available
#The code checks if the entered book title exists in the library_books dictionary.
#If the book is available and there are enough copies, one copy is deducted and a success message is displayed.
#If the book is out of stock or not available in the library, an appropriate message is displayed.

if user_book in library_books:
    if library_books[user_book] > 0:
        library_books[user_book] -= 1
        print(f"Book '{user_book}' borrowed successfully!")
    else:
        print(f"Sorry, The Book '{user_book}' is currently out of stock.")
else:
    print(f"Sorry, The Book '{user_book}' is not available in the library.")

# Display updated book availability
print("\nUpdated Available Books:")

#A loop prints the updated availability of each book.
for book, copies in library_books.items():
    print(f"{book} - Copies available: {copies}")
```

```python
#Scenario-04 : Verifying Availability of Product(s) for given Input and Update their Availability Count accordingly
#Throw the Status to user according to their Need
#Enlist the Total Cost for Added Products

store_inventory = {
    "Apples": {"Available_quantity": 100, "price": 1.5},
    "Bananas": {"Available_quantity": 50, "price": 0.8},
    "Milk": {"Available_quantity": 20, "price": 2.0},
    "Bread": {"Available_quantity": 30, "price": 1.0},
    "Eggs": {"Available_quantity": 40, "price": 1.2},
}

Shopping_cart = {}

def display_inventory():
    print("\nStore Inventory:")
    for product, details in store_inventory.items():
        print(f"{product} - Quantity: {details['Available_quantity']}, Price: ${details['price']} per unit")

display_inventory()

while True:
    product = input("\nEnter the product you want to buy (or 'Done' to finish): ").strip()
    if product.lower() == "done":
        break
    
    if product in store_inventory:
        Shopping_quantity = int(input(f"How many units of {product} do you want to buy? ").strip())
        
        if Shopping_quantity > store_inventory[product]["Available_quantity"]:
            print(f"Sorry, only {store_inventory[product]['Available_quantity']} units available.")
        else:
            store_inventory[product]["Available_quantity"] -= Shopping_quantity
            
            if product in Shopping_cart:
                Shopping_cart[product]["Cart_quantity"] += Shopping_quantity
                
            else:
                Shopping_cart[product] = {"Cart_quantity": Shopping_quantity, "price": store_inventory[product]["price"]}
            print(f"{Shopping_quantity} units of {product} added to your cart.")
    else:
        print("Product not found. Please enter a valid product name.")

print("\nYour Shopping Cart:")
total_cost = 0
for product, details in Shopping_cart.items():
    cost = details["Cart_quantity"] * details["price"]
    total_cost += cost
    print(f"{product} - Quantity: {details['Cart_quantity']}, Price per unit: ${details['price']}")
    

#total_cost:.2f (Formatting a Floating-Point Number)
#{total_cost:.2f} ensures that total_cost is displayed with two decimal places.
#:.2f means:
#: starts the formatting instruction.
#.2f means "format as a float with 2 decimal places".

print(f"\nTotal Cost of Your Purchase: ${total_cost:.2f}")
```

```python
#Scenario-05 : Create Customers with Name, A/C No. and Balance 
#Throw the Transaction Status to user (Based on Account Status) according to the Chosen Transaction
#Display updated account details after transactions for given Input

customers = {
    "Alice": {"account_number": 1001, "balance": 5000},
    "Bob": {"account_number": 1002, "balance": 3000},
    "Charlie": {"account_number": 1003, "balance": 7000}
}

def display_inventory():
    for customer, details in customers.items():
        print(f"{customer} - Account Number: {details['account_number']}, Balance: ${details['balance']}")
        
# Display initial account details
print("Initial Account Details:")
display_inventory()
    
# Second code snippet: Perform transactions (deposits and withdrawals)
for customer, details in customers.items():
    transaction_type = input(f"\n{customer}, do you want to (D)eposit or (W)ithdraw? ").upper()

    if transaction_type == 'D':
        amount = float(input("Enter the deposit amount: $"))
        details['balance'] += amount
        print(f"\nDeposit of ${amount} successful. New balance: ${details['balance']}")
    elif transaction_type == 'W':
        amount = float(input("Enter the withdrawal amount: $"))
        if amount <= details['balance']:
            details['balance'] -= amount
            print(f"\nWithdrawal of ${amount} successful. New balance: ${details['balance']}")
        else:
            print("Insufficient funds. Withdrawal unsuccessful.")
    else:
        print("Invalid transaction type. Please enter 'D' for deposit or 'W' for withdrawal.")
        
# Display updated account details after transactions
print("\nUpdated Account Details:")
display_inventory()
```

```python
#Scenario-06 : First Define Methods for Deposit and Withdrawal by Creating a Class and Define a Seperate Method to Initiate this Class
#Create Customers with Name, A/C No. and Balance 
#Throw the Transaction Status to user (Based on Account Status) according to the Chosen Transaction
#Display updated account details after transactions for given Input

#1. Defining the BankAccount Class
#Declares a class named BankAccount to represent a bank account.

class BankAccount:
   
 #2. Constructor (__init__ method)
    def __init__(self, account_number, balance): #Initializes the BankAccount object.
        self.account_number = account_number #account_number: Stores the account number.
        self.balance = float(balance) #balance: Converts the balance to a floating-point number to allow decimal values.

 #3. Deposit Method (deposit)
    def deposit(self, amount): #Takes an amount and converts it to a float.
        self.balance += float(amount) #Adds the amount to the balance.
        return f"Deposit of ${float(amount):.1f} successful. New balance: ${self.balance:.1f}"
    #Returns a formatted string showing the deposit amount and the new balance, ensuring one decimal place (.1f).

 #4. Withdraw Method (withdraw)   
    def withdraw(self, amount): 
        if float(amount) <= self.balance: #Checks if the withdrawal amount is less than or equal to the balance.
            self.balance -= float(amount) #Deducts the amount if funds are sufficient
            return f"Withdrawal of ${float(amount):.1f} successful. New balance: ${self.balance:.1f}"
            #Ensures proper decimal formatting in output.
        else:
            return "Insufficient funds. Withdrawal unsuccessful."
            #Otherwise returns an "Insufficient funds" message.
        
 #5. Creating Customer Accounts       
#Creates a dictionary called customer_accounts to store customer names as keys.
#Each customer is assigned a BankAccount object with a unique account number and initial balance.

customer_accounts = {
    "Alice": BankAccount(1001, 5000), 
    "Bob": BankAccount(1002, 3000),
    "Charlie": BankAccount(1003, 7000)
}

#6. Processing User Transactions
#Loops through each customer and their corresponding account
for customer, account in customer_accounts.items():
    
    #Prompts the user for a transaction type (D for deposit, W for withdrawal).
    #.upper() ensures the input is case-insensitive (e.g., d is treated as D).
    transaction_type = input(f"{customer}, do you want to (D)eposit or (W)ithdraw? ").upper()

#7. Handling Deposits
    if transaction_type == 'D': #If the user selects D, it asks for the deposit amount
        amount = float(input("Enter the deposit amount: $ ")) #Converts input to float and calls the deposit method.
        print(account.deposit(amount)) #Prints the deposit confirmation message.
        
#8. Handling Withdrawals
    elif transaction_type == 'W': #If the user selects W, it asks for the withdrawal amount.
        amount = float(input("Enter the withdrawal amount: $ ")) #Calls the withdraw method 
        print(account.withdraw(amount)) #prints the response.
        
#9. Handling Invalid Inputs    
    else:
        print("Invalid transaction type. Please enter 'D' for deposit or 'W' for withdrawal.")
        #Displays an error message if the user enters anything other than D or W.

#10. Displaying Updated Account Details
print("\nUpdated Account Details (using BankAccount class):")
for customer, account in customer_accounts.items():
    print(f"{customer} - Account Number: {account.account_number}, Balance: ${account.balance:.1f}")
    #Loops through each customer and prints their account number and updated balance formatted to one decimal place.
```

```python
#Scenario-07 : #Create Students with Name, Assignments Score and Final Exam Score 
#Display initial student details and assignment scores
#Calculate the average assignment score for each student
#Calculate and display final grades for each student


students = {
    "Alice": {"assignments": [90, 85, 95], "final_exam": 88},
    "Bob": {"assignments": [78, 92, 80], "final_exam": 75},
    "Charlie": {"assignments": [88, 85, 90], "final_exam": 92}
}

# Display initial student details and assignment scores
print("Student Details and Assignment Scores:")
for student, details in students.items():
    print(f"{student} - Assignments: {details['assignments']}, Final Exam: {details['final_exam']}")
    
# Second code snippet: Calculate the average assignment score for each student
for student, details in students.items():
    assignment_avg = sum(details["assignments"]) / len(details["assignments"])
    details["assignment_avg"] = assignment_avg
    
# Display updated student details with average assignment scores
print("\nUpdated Student Details with Assignment Averages:")
for student, details in students.items():
    print(f"{student} - Assignments Average: {details['assignment_avg']:.2f}, Final Exam: {details['final_exam']}")
    
# Third code snippet: Determine final grades based on assignment averages and final exam scores
def calculate_final_grade(avg_score, final_exam_score):
    final_score = 0.4 * avg_score + 0.6 * final_exam_score
    if final_score >= 90:
        return "A"
    elif final_score >= 80:
        return "B"
    elif final_score >= 70:
        return "C"
    elif final_score >= 60:
        return "D"
    else:
        return "F"

# Calculate and display final grades for each student
print("\nFinal Grades:")
for student, details in students.items():
    final_grade = calculate_final_grade(details["assignment_avg"], details["final_exam"])
    print(f"{student} - Final Grade: {final_grade}")
```
